```python
def save_target_seg_data(dataset_path, camera_names, type="train", compressed=True):
   with h5py.File(dataset_path, 'r') as root:
      for cam_name in camera_names:
         for i, image in enumerate(root[f'/observations/images/{cam_name}']):
            if compressed:
               save_path = os.path.join(target_seg_data_save, type, os.path.basename(dataset_path).replace(".hdf5", "")+ "_" + cam_name + f"_{i:05d}.jpg")
            else:
               save_path = os.path.join(target_seg_data_save, type, os.path.basename(dataset_path).replace(".hdf5", "")+ "_" + cam_name + f"_{i:05d}.png")
            print("save_path: ", save_path)
            image = root[f'/observations/images/{cam_name}'][i]
            cv.imwrite(save_path, image)
```

```python
dataset_path = dataset_path_list[0]
camera_names = ['wrist', 'wrist_down']
compressed = True

start_ts = 0
for dataset_path in train_dataset_path:
   save_target_seg_data(dataset_path, camera_names, type="train", compressed=compressed)
for dataset_path in val_dataset_path:
   save_target_seg_data(dataset_path, camera_names, type="val", compressed=compressed)
```
